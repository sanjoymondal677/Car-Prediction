# Data preprocessing

```python
def LABEL_ENCODING(c1):
    from sklearn import preprocessing
    label_encoder = preprocessing.LabelEncoder()
    data[c1]= label_encoder.fit_transform(data[c1])
    data[c1].unique()
#Converted all the object type columns to int and with help of label encoder and placed vlaues like 0,1 and antoher values
```

```python
LABEL_ENCODING("model")
LABEL_ENCODING("transmission")
LABEL_ENCODING("fuelType")
data
```

# Select the independent and dependent variables

```python
x = data.drop('price',axis=1) #Independent variables or features variables based on this features we are predict the out put
```

```python
print(data)
```

```python
y=data['price'] #Dependent variable or labeled data
```

```python
print(y)
```

# Fearures scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X1 = sc.fit_transform(x)
X1 = sc.transform(x)
```

```python
print(X1)
```

# Split our data into train and test's data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X1,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
DTR = DecisionTreeRegressor(max_depth=5)
```

```python
DTR.fit(x_train,y_train)
```

```python
DTR.score(x_train,y_train)
```

```python
dec_score = DTR.score(x_test,y_test)
print(dec_score)
```

# RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor(n_estimators=40,random_state=0)
```

```python
RF.fit(x_train,y_train)
```

```python
print(RF.score(x_train,y_train))
```

```python
ran_score = RF.score(x_test,y_test)
print(ran_score)
```
